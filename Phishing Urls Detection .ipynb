```python
listing2= listing1.drop(columns={'nb_dollar','nb_space','nb_www','nb_com','nb_dslash','http_in_path','https_token','ratio_digits_url','ratio_digits_host','ratio_intMedia','ratio_extMedia','sfh','iframe','punycode','port','tld_in_path','tld_in_subdomain','abnormal_subdomain','nb_subdomains','prefix_suffix','random_domain','shortening_service','login_form','external_favicon','submit_email','domain_registration_length','path_extension','nb_redirection','nb_external_redirection','length_words_raw','char_repeat','shortest_words_raw','shortest_word_host','shortest_word_path','longest_words_raw','ratio_intRedirection','ratio_extRedirection','ratio_intErrors','ratio_extErrors','longest_word_host','longest_word_path','avg_words_raw','avg_word_host','avg_word_path','phish_hints','domain_in_brand','brand_in_subdomain','brand_in_path','ratio_intHyperlinks','ratio_extHyperlinks','ratio_nullHyperlinks','nb_extCSS','suspecious_tld','statistical_report'})
```

```python
listing2
```

```python
listing3=listing2.rename(columns={'url':'Urls','nb_hyperlinks':'Hyperlinks','links_in_tags':'Links in Tags','safe_anchor':'Safe Anchor','domain_with_copyright':'Domain with Copyright','web_traffic':'Web Traffic','page_rank':'Page Rank'})	
```

```python
listing3
```

```python
listing4= listing3.drop_duplicates()
```

```python
listing4
```

```python
listing4.loc[:, 'Safe Anchor'] = listing4['Safe Anchor'].round(2)
```

```python
listing4.loc[:, 'Links in Tags'] = listing4['Links in Tags'].round(2)
```

```python
listing4
```

```python
listing4['legitimate_count'] = listing4['status'].apply(lambda x: 1 if x.lower() == 'legitimate' else 0)
```

```python
legitimate_count = (listing4['status'].str.lower() == 'legitimate').sum()
```

```python
phishing_count = (listing4['status'].str.lower() == 'phishing').sum()
```

```python
average_safe_anchor = listing4['Safe Anchor'].mean()
```

```python
average_links_in_tags = listing4['Links in Tags'].mean()
```

```python
KPIS = pd.DataFrame({
    'URLs': ['KPI','KPI','KPI','KPI'],
    'status': ['Phishing Count', 'Legitimate Count', 'Average Safe Anchor', 'Average Links in Tags'],
     'Safe Anchor': [None, None, average_safe_anchor, None],
     'Links in Tags': [None, None, None, average_links_in_tags],
    'Value': [phishing_count, legitimate_count, None, None]
})
```

```python
KPIS
```

```python
KPIS.to_csv('KPIS_for_PowerBI.csv', index=False)
```
